```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import (
    accuracy_score,
    recall_score,
    precision_score,
    classification_report
)
```

```python
def evaluate_model(y_true, y_pred):
    """Evaluate model performance and print classification report."""
    accuracy = accuracy_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)

    print(f"\nModel Evaluation:")
    print(f"Accuracy Score: {accuracy * 100:.2f}%")
    print(classification_report(y_true, y_pred))

    return accuracy, recall, precision
```

```python
def extract_features_and_labels(dataframe):
    """Extract features and target variable from dataframe."""
    features = dataframe[["x", "y"]].values
    labels = dataframe["result"].values
    return features, labels
```

```python
def train_dummy_classifier(X_train, y_train, X_test, y_test):
    """Train and evaluate Dummy Classifier."""
    dummy_model = DummyClassifier(strategy="stratified")
    dummy_model.fit(X_train, y_train)
    y_pred = dummy_model.predict(X_test)

    accuracy, recall, precision = evaluate_model(y_test, y_pred)
    return dummy_model, accuracy, recall, precision
```

```python
def train_knn_classifier(X_train, y_train, X_test, y_test, k=15):
    """Train and evaluate K-Nearest Neighbors Classifier."""
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(X_train, y_train)
    y_pred = knn_model.predict(X_test)

    accuracy, recall, precision = evaluate_model(y_test, y_pred)
    return knn_model, accuracy, recall, precision
```

```python

def run_batch_model_pipeline(file_path1, file_path2):
    """Run a stepwise training, testing, and evaluation of models."""
    results = {}

    # Step 1: Initial dataset processing
    df1 = pd.read_csv(file_path1)
    X1, y1 = extract_features_and_labels(df1)
    X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.40, random_state=42)

    # Step 2: Dummy Classifier
    dummy_model, acc, rec, prec = train_dummy_classifier(X_train1, y_train1, X_test1, y_test1)
    results["Step 1: Dummy Classifier"] = {"accuracy": acc, "recall": rec, "precision": prec}

    # Step 3: Train KNN on subset
    X_train_sub, X_eval, y_train_sub, y_eval = train_test_split(X_train1, y_train1, test_size=0.40, random_state=42)
    knn_model, acc, rec, prec = train_knn_classifier(X_train_sub, y_train_sub, X_eval, y_eval)
    results["Step 2: KNN Initial"] = {"accuracy": acc, "recall": rec, "precision": prec}

    # Step 4: KNN Predictions on initial test set
    y_pred1 = knn_model.predict(X_test1)
    acc, rec, prec = evaluate_model(y_test1, y_pred1)
    results["Step 3: KNN Predict Initial"] = {"accuracy": acc, "recall": rec, "precision": prec}

    # Step 5: Load second dataset and retrain
    df2 = pd.read_csv(file_path2)
    X2, y2 = extract_features_and_labels(df2)
    X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.40, random_state=42)
    X_train_sub2, X_eval2, y_train_sub2, y_eval2 = train_test_split(X_train2, y_train2, test_size=0.40, random_state=42)

    knn_model, acc, rec, prec = train_knn_classifier(X_train_sub2, y_train_sub2, X_eval2, y_eval2)
    results["Step 4: KNN Retrained"] = {"accuracy": acc, "recall": rec, "precision": prec}

    # Step 6: Predictions on second dataset test set
    y_pred2 = knn_model.predict(X_test2)
    acc, rec, prec = evaluate_model(y_test2, y_pred2)
    results["Step 5: KNN Predict Retrained"] = {"accuracy": acc, "recall": rec, "precision": prec}

    return results
```

```python
def results_to_dataframe(results_dict):
    """Convert results dictionary to pandas DataFrame."""
    results_list = []
    for step, metrics in results_dict.items():
        metrics["stage_name"] = step
        results_list.append(metrics)
    return pd.DataFrame(results_list)
```

```python
def plot_metrics(results_df):
    """Plot accuracy, recall, and precision across stages."""
    plt.figure(figsize=(10, 6))
    plt.plot(results_df["stage_name"], results_df["accuracy"], marker='o', label='Accuracy', linewidth=2)
    plt.plot(results_df["stage_name"], results_df["recall"], marker='o', label='Recall', linewidth=2)
    plt.plot(results_df["stage_name"], results_df["precision"], marker='o', label='Precision', linewidth=2)

    plt.title("Model Performance Across Steps")
    plt.xlabel("Model Stages")
    plt.ylabel("Metric Scores")
    plt.xticks(rotation=30)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
dataset_path_1 = "first.csv"
dataset_path_2 = "second.csv"

pipeline_results = run_batch_model_pipeline(dataset_path_1, dataset_path_2)
results_df = results_to_dataframe(pipeline_results)
print(results_df)

plot_metrics(results_df)
```
